# Precessão do periélio de Mercúrio

As órbitas dos planetas ao redor do sol, apresentam pequenas variações no caminho previsto. O planeta segue uma elipse, e essa rotação da órbita é chamada de **precessão**.

4al3U.png


##Desvios inexplicáveis ​​na órbita de Mercúrio
Em 1859, o astrônomo francês, Urbain Le Verrier, ao examinar dados antigos -percebeu a própria órbita de Mercúrio estava gradualmente girando a cada revolução que passava.

A teoria de Newton prevê esses efeitos de precedência, como sendo produzidos pela atração dos planetas uns sobre os outros. 

Porém, a precessão esperada das apsides de Mercúrio, é de 531 segundos de arco por século, e observaram que a precessão real é de 574 segundos de arco por século. Isso deixa uma diferença de 43" que não pode ser explicada por incertezas no cálculo ou na medição.




mercurio.jpg


Então, para justificar essa discrepância pela Lei da gravitação universal, a mecânica clássica de Newton, supôs que poderia haver um planeta não descoberto que Le Verrier chamou de "Vulcano" dentro da órbita de Mercúrio que estava distorcendo silenciosamente sua órbita.

```python
Web VPython 3.2
### Dados 
G = 6.67e-11 ### Constante 
msol = 1.98892e30 ### Massa do sol
Rsol = 696340000 ### Raio do sol

### Background
###scene.autoscale = False
sphere(pos=vector(0,0,0),texture="https://i.imgur.com/vws7IzD.jpeg",radius=1500*Rsol,shininess=0)
#scene.camera.pos = vector(0,0,1000*Rsol)

##### massa dos planetas
mmercurio = 3.285e23  ### Massa de mercurio 
mvenus = 4.867e26
mterra = 5.972e24 
mmarte = 6.39e23 
mjupiter = 1.898e27 
msaturno = 5.683e26
mvulcano = 3e23

#### velocidade dos planetas
vmercurio = vector(0,25e3,0) 
vvenus = vector(0,25e3,0) 
vterra = vector(0,25e3,0)
vmarte = vector(0,24e3,0)
vjupiter = vector(0,22e3,0)
vsaturno = vector(0,23e3,0)
vvulcano = vector(0,30e3,0)

### SOL 

sol = sphere(pos=vector(0,0,0), radius=10*Rsol, texture="https://i.imgur.com/PWnSJQd.jpeg")
sol.v = vector(0,0,0)
sol.p = msol*sol.v

### PLANETAS
mercurio = sphere(pos=vector(150*Rsol,0,0), radius=2*Rsol, texture = "https://i.imgur.com/SLgVbwD.jpeg", make_trail=True)
mercurio.p = mmercurio*vmercurio

venus = sphere(pos=vector(200*Rsol,0,0), radius=3*Rsol, texture = "https://i.imgur.com/7VTEX2w.jpeg", make_trail=True)
venus.p = mvenus*vvenus

terra = sphere(pos=vector(250*Rsol,0,0), radius=5*Rsol, texture="https://i.imgur.com/ns7Q3s6.jpeg")
terra.p = mterra*vterra

marte = sphere(pos=vector(300*Rsol,0,0), radius=4.5*Rsol, texture = "https://i.imgur.com/Mwsa16j.jpeg")
marte.p = mmarte*vmarte

jupiter = sphere(pos=vector(350*Rsol,0,0), radius=6*Rsol, texture = "https://i.imgur.com/RMMtt0K.jpeg")
jupiter.p = mjupiter*vjupiter

saturno = sphere(pos=vector(400*Rsol,0,0), radius=7*Rsol, texture = "https://i.imgur.com/r9h0U9E.jpeg")
saturno.p = msaturno*vsaturno

vulcano = sphere(pos=vector(100*Rsol,0,0), radius=2*Rsol,texture = "https://i.imgur.com/3KI442U.png", make_trail=True)
vulcano.p = mvulcano*vvulcano

### LAÇO
t = 0
dt = 20
tf = 70000
while t<tf:
  rate(150)

  ### Vector r
  rms = mercurio.pos - sol.pos
  
  ### Forca sobre o satélite 

  F1= ((-G*msol*mmercurio)/(mag(rms)**2))*norm(rms) 
  ### Momento Linear do satélite 
  
  mercurio.p = mercurio.p + (F1*t)
  
  ### Posição do satelite 

  mercurio.pos = mercurio.pos + (mercurio.p/mmercurio)*t 
  
  ##### corpo 2
  ### Vector r
  rvs = venus.pos - sol.pos
  ### Forca sobre o satélite 

  F2 = ((-G*msol*mvenus)/(mag(rvs)**2))*norm(rvs) 

  ### Momento Linear do satélite 
  
  venus.p = venus.p + (F2*t)
  
  ### Posição do satelite 

  venus.pos = venus.pos + (venus.p/mvenus)*t
  
  ### corpo 3 
  ### Vector r
  rts = terra.pos - sol.pos
  ### Forca sobre o satélite 

  F3 = ((-G*msol*mterra)/(mag(rts)**2))*norm(rts) 

  ### Momento Linear do satélite 
  
  terra.p = terra.p + (F3*t)
  
  ### Posição do satelite 

  terra.pos = terra.pos + (terra.p/mterra)*t
  
  ### corpo 4
  ### Vector r
  rmas = marte.pos - sol.pos
  ### Forca sobre o satélite 

  F4 = ((-G*msol*mmarte)/(mag(rmas)**2))*norm(rmas) 

  ### Momento Linear do satélite 
  
  marte.p = marte.p + (F4*t)
  
  ### Posição do satelite 

  marte.pos = marte.pos + (marte.p/mmarte)*t

 ### corpo 5
  ### Vector r
  rjs = jupiter.pos - sol.pos
  ### Forca sobre o satélite 

  F5 = ((-G*msol*mjupiter)/(mag(rjs)**2))*norm(rjs) 

  ### Momento Linear do satélite 
  
  jupiter.p = jupiter.p + (F5*t)
  
  ### Posição do satelite 

  jupiter.pos = jupiter.pos + (jupiter.p/mjupiter)*t
  
  ### corpo 6
  ### Vector r
  rss = saturno.pos - sol.pos
  ### Forca sobre o satélite 

  F6 = ((-G*msol*msaturno)/(mag(rss)**2))*norm(rss) 

  ### Momento Linear do satélite 
  
  saturno.p = saturno.p + (F6*t)
  
  ### Posição do satelite 

  saturno.pos = saturno.pos + (saturno.p/msaturno)*t
  
  ######### Vulcano
  rvvv = vulcano.pos - sol.pos
  ### Forca sobre o satélite 

  F00 = ((-G*msol*mvulcano)/(mag(rvvv)**2))*norm(rvvv) 

  ### Momento Linear do satélite 
  
  vulcano.p = vulcano.p + (F00*t)
  
  ### Posição do satelite 

  vulcano.pos = vulcano.pos + (vulcano.p/mvulcano)*t
  
  
  ### Vector r
  rms = mercurio.pos - sol.pos
  rmv = mercurio.pos - venus.pos
  rmt = mercurio.pos - terra.pos
  rmma = mercurio.pos - marte.pos
 
  
  t += dt
```
